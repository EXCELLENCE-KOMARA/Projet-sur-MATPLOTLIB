```python
#remplacer les valeurs nulles par 0
df_process = df.fillna(0)
```

```python
#moyenne de prix de vente en fonction de  l'année de fabrication
plt.figure(figsize=(8,5))
df_process.groupby('year')['selling_price'].mean().plot(marker = 'o', color = 'magenta')
plt.title('Année de fabrication et prix de vente')
plt.xlabel('Année de fabrication')
plt.ylabel('moyenne de prix de vente')
plt.grid(True)
plt.savefig('mon_graphique.png', dpi = 300)
plt.show()
```

```python
#histogramme de distribution de prix de vente
plt.figure(figsize=(8,5))
plt.hist(df_process['selling_price'],color = 'magenta', edgecolor= 'yellow', bins =40)
plt.title('distribution de prix de vente')
plt.xlabel('prix de vente')
plt.ylabel('nombre de véhicule')
plt.grid()
plt.savefig('mon_graphique1',dpi=400)
plt.show()
```

```python
#calucul des moyenne de prix de vente des véhicules
df_mean_price = df_process.groupby('fuel')['selling_price'].mean().reset_index()
df_mean_price= df_mean_price.rename(columns={'selling_price':'prix moyen vente'})
df_mean_price
```

```python
#  prix moyen  par catégorie  de carburant
plt.figure(figsize=(10,5))
sns.barplot(data=df_mean_price, x= 'fuel', y = 'prix moyen vente')
plt.title('Prix moyen par  catégorie de carburant')
plt.xlabel('carburant')
plt.ylabel('prix moyen')
plt.grid('fuel')
plt.savefig('mon_graphique4', dpi= 300)
plt.show()
```

```python
#rélation entre prix et kilomère parcouri par type de carburant
plt.figure(figsize=(8,5))
sns.scatterplot(data=df_process, x='km_driven', y= 'selling_price'  , color = 'green', hue= 'fuel')
plt.title('relation prix et kilomètre')
plt.xlabel('kilomèrre')
plt.ylabel('prix')
plt.grid()
plt.savefig('mon_graphique', dpi=300)
plt.show()
```

```python
#correlaton entre l'année de fabrication et prix de vente
plt.figure(figsize = (8,5))
sns.regplot(data =df_process, x = 'year' , y = 'selling_price', marker = '*', scatter_kws ={'color':'cyan'},line_kws={'color':'orange'} )
plt.title("l'année de fabrication et  prix de vente ")
plt.xlabel('Année')
plt.ylabel('prix')
plt.grid(True)
plt.savefig('us_graphique.png', dpi= 400)
plt.show()
```

```python
#rélation entre entre l'année de fabrication, kilomère parcouri et prix de vente
x = df_process['km_driven']
y = df_process['selling_price']
z = df_process['year']
#création de la figure et du plot 3D
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')
#nuage de points
ax.scatter(x, y, z, c= 'green', marker='o')
#les étiquettes
ax.set_xlabel('kilomètrage')
ax.set_ylabel('Prix de vente')
ax.set_zlabel('Année')
plt.title('visualisation en 3D des Véhicules')
plt.show()
```

```python
import numpy as np

# Supposons que df_process contient : year, km_driven, selling_price

# Prendre un échantillon (bar3d est lent si trop de données)
df_bar = df_process.sample(100).copy()

# Normaliser ou transformer les données pour avoir des tailles comparables
df_bar['year_norm'] = df_bar['year'] - df_bar['year'].min()
df_bar['km_norm'] = df_bar['km_driven'] / 10000  # Échelle plus compacte
df_bar['price_norm'] = df_bar['selling_price'] / 10000

# Position de départ des barres
x = df_bar['year_norm']
y = df_bar['km_norm']
z = np.zeros_like(x)

# Dimensions des barres
dx = dy = 0.5  # largeur et profondeur fixes
dz = df_bar['price_norm']  # hauteur = prix

# Création de la figure
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Affichage des barres 3D
ax.bar3d(x, y, z, dx, dy, dz, color='steelblue', alpha=0.8)

# Étiquettes
ax.set_xlabel("Année (normalisée)")
ax.set_ylabel("Kilométrage (normalisé)")
ax.set_zlabel("Prix (en 10 000)")
plt.title("Barres 3D : Prix de vente en fonction de l'année et du kilométrage")

plt.show()
```
